```python
# Utils
import sys
import os
import shutil
import time
import pickle
import numpy as np

# Deep Learning Librairies
import tensorflow as tf
import keras.preprocessing.image as kpi

import keras.layers as kl
from keras.models import Sequential
from keras.layers import Dense
import keras.optimizers as ko
import keras.backend as k
import keras.models as km
import keras.applications as ka

# Visualisaiton des données
from matplotlib import pyplot as plt
```

```python
# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Debut du decompte du temps
import time
start_time = time.time()
```

```python
# data, x,y
x = np.arange(0, np.pi * 2, 0.1)
y = np.sin(x)

# hyperparametres
ACTIVE_FUN = 'tanh'
BATCH_SIZE = 1
VERBOSE = 0

# architecture RN - couche dense
model = Sequential()
model.add(Dense(5, input_shape=(1,), activation=ACTIVE_FUN))
model.add(Dense(5, activation=ACTIVE_FUN))
model.add(Dense(1, activation='linear'))

# modele + rmse + optimiseur
model.compile(loss='mean_squared_error', optimizer='sgd', metrics=['mean_squared_error'])

# Train - Y 
model.fit(x, y, epochs=1000, batch_size=BATCH_SIZE, verbose=VERBOSE)

# Perte
scores = model.evaluate(x, y, verbose=VERBOSE)
print('%s: %.2f%%' % (model.metrics_names[1], scores[1] * 100))

# Test - Ychapeau
y_pred = model.predict(x)
```

```python
# Analyse : mean_squared_error: 0.01%
```

```python
# Graphique
plt.plot(x, y, color='blue', linewidth=1, markersize='1')
plt.plot(x, y_pred, color='green', linewidth=1, markersize='1')
plt.xlabel('Angle [rad]')
plt.ylabel('y=sin(x)')
plt.axis('tight')

# Graphique lisible en N&B
fig, ax = plt.subplots()
ax.plot(x, y, 'k--', label='y : blue')
ax.plot(x, y_pred, 'k:', label='ychapeau : vert')
legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
plt.xlabel('Angle [rad]')
plt.ylabel('y=sin(x)')
plt.legend()
plt.show()
```
